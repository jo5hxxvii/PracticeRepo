```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
## kind can be = hex, reg, kde...
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

  # Categorical Data

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
flights = sns.load_dataset('flights')
tipcorr = tips.corr()
tipcorr
```

```python
flights.head()
```

```python
sns.heatmap(tipcorr, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns = 'year', values='passengers')
```

```python
##sns.heatmap(fp, cmap='coolwarm', linewidths=1, linecolor='white')
sns.heatmap(fp)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', 
           col='day', aspect=0.6, size=8)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
fg = sns.FacetGrid(data=tips,col='time',row='smoker')
fg.map(plt.scatter,'total_bill','tip')
```
